```python
def plot_ad(ad):
    plt.figure(figsize=(5,5))

    atac_metacells = ad.obs['Metacell'].unique()
    metacell_colors = pd.Series(sns.color_palette(n_colors=len(atac_metacells)).as_hex(), 
             index=np.random.choice(atac_metacells, len(atac_metacells), False))

    basis = 'X_umap'

    plt.scatter(ad.obsm[basis][:, 0], 
                -ad.obsm[basis][:, 1], s=.05, 
               color=metacell_colors[ad.obs['Metacell']])#, edgecolor='black')

    df = pd.DataFrame(ad[list(atac_metacells)].obsm[basis], index=atac_metacells)
    df = pd.DataFrame(ad.obsm[basis]).set_index(ad.obs_names).join(ad.obs['Metacell']).groupby('Metacell').mean()
    plt.scatter(df[0], -df[1], s=20, 
               color=metacell_colors[df.index], linewidth=0.25,
                edgecolor='black', alpha=1)

    ax = plt.gca()
    ax.set_axis_off()
    plt.show()
    plt.close()
```

```python
from src.metacells.core import Metacells

for file in to_process[:3]:
    
    print(base_dir+file)
    ad = sc.read(base_dir + file)
    ad.obs.rename(columns={'CELL_TYPE':'celltype'}, inplace=True)
    
    if 'celltype' not in ad.obs:
        ad.obs['celltype'] = 'unknown'
        
    if 'atac' in file:
        build_on = "X_svd"
        dimension = 'SVD'
    else:
        build_on = "X_pca"
        dimension = "PCA"
        
    print(file)
    
    # Grab Tanay Labels as well... 
    tanay_dir = base_dir+''.join(file.split('.')[:-1])+'_tanay/'
    results_dir = base_dir+''.join(file.split('.')[:-1])+'_results/'
    
    os.makedirs(results_dir, exist_ok=True)
    
    print('Reading tanay files from ',tanay_dir)
    
    # Read in the metacells and assignments from Tanay method
    assmts = pd.read_csv(tanay_dir+'metacells.txt', sep=" ")
    assmts.index = assmts.index.astype(str)

    label_df = assmts.rename(columns = {'x':'tanay_labels'})
    label_df.index = label_df.index.astype(str) 
    n_metacells = len(label_df['tanay_labels'].unique())
    
    print(f'Original dataset contained {ad.n_obs} cells. Tanay kept {assmts.shape[0]} = {(100*assmts.shape[0]/ad.n_obs):2f}%')
    
    if keep_tanay_cells_only:
        print('Subsetting Tanay Cells')
        init_ad = ad 
        ad = bm.subset_anndata(ad[label_df.index, :])
        
    model = Metacells(ad, 
                      build_kernel_on=build_on, 
                      n_metacells=n_metacells, 
                      n_waypoint_eigs=10)

    model.fit(n_iter=8)
    
    ad.write(base_dir+file.strip('.h5ad')+'_compare_tanay.h5ad')
    
    plot_ad(ad, save_as=base_dir+file.strip('.h5ad')+'2D.png')
    
    tanay_ad = bm.subset_anndata(ad.copy())
    tanay_ad.obs = ad.obs.drop('Metacell', axis=1).join(assmts.rename(columns={'x':'Metacell'}))
    
    plot_ad(tanay_ad, save_as=base_dir+file.strip('.h5ad')+'2D_tanay.png')

    
    
```

```python
from src.metacells import evaluate

processed = ['cd34_multiome_atac_no_bcells.h5ad', 'pbmc_multiome_atac.h5ad', 'E8.75_endoderm_atac.h5ad'] #['cd34_multiome_rna_no_bcells.h5ad','cd34_multiome_atac_no_bcells.h5ad','lung_tumor_non_immune.h5ad', 'lung_tumor_immune.h5ad']

processed = ['pbmc_multiome_rna.h5ad', 'cd34_multiome_rna_no_bcells.h5ad', 'lung_tumor_non_immune.h5ad', 'lung_tumor_immune.h5ad']
nbr = 1

compactness = {'compactness':[], 'method':[], 'dataset':[]}
separation = {'separation':[],'separation_filtered':[], 'method':[], 'dataset':[]}


for file in processed:
    ad = sc.read(base_dir+file.strip('.h5ad')+'_compare_tanay.h5ad')
    
    tanay_dir = base_dir+''.join(file.split('.')[:-1])+'_tanay/'
    
    print('Reading tanay files from ',tanay_dir)
    # Read in the metacells and assignments from Tanay method
    assmts = pd.read_csv(tanay_dir+'metacells.txt', sep=" ")
    assmts.index = assmts.index.astype(str)
    
    tanay_ad = bm.subset_anndata(ad.copy())
    tanay_ad.obs = ad.obs.drop('Metacell', axis=1).join(assmts.rename(columns={'x':'Metacell'}))
    
    plot_ad(ad)
    plot_ad(tanay_ad)

    nbrs, nlls = evaluate.get_NLLs(ad, key='X_pca')
    nbrs_f, nlls_f = evaluate.filter_neighbours(nbrs, nlls, cluster='celltype')

    tanay_nbrs, tanay_nlls = evaluate.get_NLLs(tanay_ad, key='X_pca')
    tanay_nbrs_f, tanay_nlls_f = evaluate.filter_neighbours(tanay_nbrs, tanay_nlls, cluster='celltype')
   

    # Compare tanay compactness against our compactness
    x = nlls.groupby('mc_id').mean()['self']
    compactness['compactness'] += x.values.tolist()
    compactness['method'] += ['metacells']*len(x)
    compactness['dataset'] += [file.strip('.h5ad')]*len(x)
    
    x = tanay_nlls.groupby('mc_id').mean()['self']
    compactness['compactness'] += x.values.tolist()
    compactness['method'] += ['tanay']*len(x)
    compactness['dataset'] += [file.strip('.h5ad')]*len(x)
    
    
    # Compare tanay separation against our separation 
    nlls_f['sep'] = nlls_f[nbr]-nlls_f['self']
    x = nlls_f.groupby('mc_id').mean()['sep'].values.tolist()
    separation['separation_filtered'] += x
    
    nlls['sep'] = nlls[nbr]-nlls['self']
    x = nlls.groupby('mc_id').mean()['sep'].values.tolist()
    separation['separation'] += x
    
    separation['method'] += ['metacells']*len(x)
    separation['dataset'] += [file.strip('.h5ad')]*len(x)
    
    tanay_nlls_f['sep'] = tanay_nlls_f[nbr]-tanay_nlls_f['self']
    x = tanay_nlls_f.groupby('mc_id').mean()['sep'].values.tolist()
    separation['separation_filtered'] += x
    
    tanay_nlls['sep'] = tanay_nlls[nbr]-tanay_nlls['self']
    x = tanay_nlls.groupby('mc_id').mean()['sep'].values.tolist()
    separation['separation'] += x
    
    separation['method'] += ['tanay']*len(x)
    separation['dataset'] += [file.strip('.h5ad')]*len(x)
    
```

```python
%matplotlib inline

# Plot compactness as boxplot
sns.set_style("white")
compactness = pd.DataFrame(compactness)
compactness.to_csv('rna_compactness_comparison.csv')
plt.figure(figsize=(10,10))
sns.boxplot(x='dataset', hue='method', y='compactness', data=compactness)
plt.xticks(rotation=90)
plt.show()
plt.savefig('rna_compactness_comparison.png', dpi=150)
plt.close()
```

```python
# Plot separation as boxplot
plt.figure(figsize=(10,10))
separation = pd.DataFrame(separation)
separation.to_csv('rna_separation_comparison.csv')
sns.boxplot(x='dataset', hue='method', y='separation_filtered', data=separation)
plt.xticks(rotation=90)
plt.ylim(-10,500)
plt.show()
plt.savefig('rna_separation_comparison.png')
plt.close()
```
